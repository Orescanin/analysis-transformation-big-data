Ucitavamo transformisane podatke i prikazujemo na primer prosecan winrate svakog tima u reg sezoni(izmedju 2000-2021 god)

```
from datetime import datetime, date
import pandas as pd
from pyspark.sql import Row, Column
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```
# File location and type
file_location = "/FileStore/tables/transformed_nbaReg.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df1 = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

#display(df1.orderBy())
display(df1)
```

Isto prosecan winrate samo za playoff

```
# File location and type
file_location = "/FileStore/tables/transformed_nbaPoff.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df2 = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df2)
```

## Hipoteza1:
  Klubovi u playoff-u generalno(tokom svih sezona, to jest 21 koju imamo) imaju manji winrate(procenat pobede) u odnosu na winrate u regularnom delu sezone,
  logicno jer igraju samo protiv najboljih timova
  
  
-Proveravamo hipotezu tako sto uzimamo prosecan winrate tokom svih sezona u regularnom delu i oduzmemo od prosecnog winrate-a u playoffu

```
#regWin = df1.select('tim','WIN%').groupBy('tim').avg('WIN%').alias("Reg_win%")
regWin = df1.groupBy('team').agg(avg('win_percentage').alias("Reg_win_avg")).orderBy("Reg_win_avg",ascending = False)
regWin = regWin.withColumn('Reg_win_avg',round(regWin.Reg_win_avg,2))
display(regWin)

#poffWin = df2.select('team','win_percentage').groupBy('team').avg('win_percentage').alias("Poff_win%")
poffWin = df2.groupBy('team').agg(avg('win_percentage').alias("Poff_win_avg")).orderBy("Poff_win_avg",ascending = False)
poffWin = poffWin.withColumn('Poff_win_avg',round(poffWin.Poff_win_avg,2))
display(poffWin)

dfJoin = regWin.join(poffWin, regWin.team == poffWin.team, "inner")
dfJoin = dfJoin.drop(poffWin.team)
dfJoin = dfJoin.select('*',(dfJoin.Poff_win_avg - dfJoin.Reg_win_avg).alias("divergence"))
dfJoin = dfJoin.withColumn('divergence',round(dfJoin.divergence,2))


display(dfJoin)
```

Vidimo da je hipoteza skoro tacna ali ne i potpuno, jer ima nekoliko timova(od 37 ukupno) koji zapravo vise pobedjuju u playoffu

```
display(dfJoin)
```

## Hipoteza2: 
 Klubovi koji najcesce prolaze u playoff imaju i najvise osvojenih prstenja


  -Brojiimo koliko je puta tim ucestvovao u regularnoj sezoni i delimo taj broj sa brojem koliko je puta ucestvovao u playoffu.
  Posle samo pogledamo timove koji imaju  najvise prstenja

```
df1Reg = df1.groupBy("team").agg(count("team").alias("regular")).orderBy("regular",ascending = False)
display(df1Reg)

df2App = df2.groupBy("team").agg(count("team").alias("appearance")).orderBy("appearance",ascending = False)
display(df2App)

dfJoined = df1Reg.join(df2App, df1Reg.team == df2App.team, "inner")
dfJoined = dfJoined.drop(df2App.team)
dfJoined = dfJoined.select('*', (dfJoined.appearance / dfJoined.regular).alias("prosek_kvalifikovanja"))
dfJoined = dfJoined.withColumn('prosek_kvalifikovanja',round(dfJoined.prosek_kvalifikovanja,2))


display(dfJoined)
```

```
poffWinnersSeasons = df2.select("team","season").where(df2.wins==16).orderBy('season', ascending = False)
display(poffWinnersSeasons)
poffWinnersRings = poffWinnersSeasons.groupBy("team").agg(count("team").alias("prstenje")).orderBy("prstenje",ascending = False)

display(poffWinnersRings)
```

Vidimo da je opet hipoteza uglavnom tacna, osim anomalije koja je Golden State Warriors uglavnom timovi koji imaju najvise prstenja
jesu u top 10 po proseku kvalifikovanja (grafovi iz nekog razloga nece da se prikazu samo pojave na sekundu i nestanu)

```
display(dfJoined)
display(poffWinnersRings)
```

##Hipoteza3:
  Sezona kada su klubovi osvojili prsten jesu one kada im je bio najbolji timski box +/- (objasnjenje: https://www.breakthroughbasketball.com/stats/definitions.html#:~:text=%2B%2F%2D%20%E2%80%93%20the%20plus%2Fminus%20statistic,the%20player%20exits%20the%20game.) U regularnoj sezoni.
  
Samo uzimamo godine kada su timovi osvojili prsten i proveravamo da li im je +/- te godine bio jedan od najvecih

```
display(poffWinnersSeasons)
plusminusTeam = df1.groupBy("team").agg(max("plus_minus").alias("najobolji_eff")).orderBy("najobolji_eff",ascending = False)
dfJoined = poffWinnersSeasons.join(plusminusTeam, poffWinnersSeasons.team == plusminusTeam.team, "inner")
#plusminusSez = df1.groupBy("team").agg(max("plus_minus").alias("najobolji_eff")).orderBy("najobolji_eff",ascending = False)
#plusminus = df1.select("team","season","plus_minus")
display(plusminusTeam)
display(dfJoined)
```

Uvidjamo da je hipoteza zapravo tacna svake godine kada su timovi imali svoj maksimalan efficency rating (box +/-) te godine su i osvojili prsten (opet problem sa grafovima)
